```python
# getting the time

def get_date(dateString):
  from dateutil.parser import parse
  from datetime import datetime
  theDate = parse(dateString)
  #print(theDate)
  
  # Return a format that can be easily sorted. 
  return "{:%Y-%m-%d}".format(theDate)
```

### Plotting events in time period (Morning/Afternoon/Night) of each day

```python
import time
def getHour(String):
  ts = time.strptime(String[:19], "%Y-%m-%dT%H:%M:%S")
  return int(time.strftime("%H", ts))

```

```python
getHour(edic[0]['created_at'])
```

```python
def classifyTime(hour):
  if hour >= 6 and hour <= 11:
    return "Early Bird"
  elif hour <= 17 and hour >= 12: # 17H is 5pm
    return "Midday"
  else:
    return "Night Owl"
```

```python
classifyTime(getHour(edic[0]['created_at']))
```

```python
workingTime = {'Early Bird': {}, 'Midday': {}, 'Night Owl': {}}

for number in edic:
  t = getHour(edic[number]['created_at'])
  date = get_date(edic[number]['created_at'])
  if classifyTime(t) == 'Early Bird':
    if date not in workingTime['Early Bird'].keys():
      workingTime['Early Bird'][date] = 1
    else:
      workingTime['Early Bird'][date] += 1
  elif classifyTime(t) == 'Midday':
    if date not in workingTime['Midday'].keys():
      workingTime['Midday'][date] = 1
    else:
      workingTime['Midday'][date] += 1
  else:
    if date not in workingTime['Night Owl'].keys():
      workingTime['Night Owl'][date] = 1
    else:
      workingTime['Night Owl'][date] += 1
      
print(workingTime)
```

```python
# formatData function which takes data in the input data format in the form of a list of dictionaries with two keys x 
# and y. It then converts it into the required format required by the charts CodeSandbox

# Input data format [line,pie or bar charts]:
# -line: (x,y) - coordinates of points ordered in the list sequentially
# -bar: x - name of category, 
#       y - count of specific categories
# -pie: x - name of category
#       y - count of specific category
#
# style = recharts or victory
# chartType = line, pie or bar
# title = title of chart
# yaxisLabel = y axis label of chart
# xaxisLabel = x axis label of chart

# Note: Data edited and converted into a dictionary. Copied from jupyter notebook.

def formatData(data,style,chartType,title,yaxisLabel,xaxisLabel):
    ans = {}
    ans['data']=list(data)
    ans['style'] = style
    ans['chartType'] = chartType
    ans['title'] = title
    ans['yaxisLabel'] = yaxisLabel
    ans['xaxisLabel'] = xaxisLabel
    
    if style=='victory':
        if chartType == 'line' or chartType =='pie' or chartType == 'bar':
            return ans
        else:
            return {}

    elif style=='recharts':
        if chartType == 'line':
            return ans
        elif chartType == 'pie':
            pieData = []
            for d in data:
                x = d['x']
                y = d['y']
                pieData.append({'name':x, 'value':y})
            ans['data'] = pieData
            return ans
        elif chartType == 'bar':
            return ans
    else:
        return 'ERROR'
```
